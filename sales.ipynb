```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
import lightgbm as lgbm
import keras 
from keras.models import Sequential
from keras.layers import Dense
```

```python
train= pd.read_csv('rr_train.csv')
test= pd.read_csv('rr_test.csv')
sub = pd.read_csv('sample_submission.csv')
```

```python
for columns in train.columns:
    print(train[columns].value_counts())
```

```python
for columns in test.columns:
    print(test[columns].value_counts())
```

```python
trainX = train.drop(['revenue'],axis=1)
trainY = train['revenue']
```

```python
model = Sequential()
model.add(Dense(16,activation='relu',input_dim=trainX.shape[1]))
model.add(Dense(8,activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error',optimizer='adam')

model.fit(trainX,trainY,verbose=2,epochs=1000)
```

```python
testX= test.drop(['ID'],axis=1)
testY = test['ID']
```

```python
xgb1 = xgb.XGBRegressor(colsample_bytree=0.4,
                 gamma=0,                 
                 learning_rate=0.001,
                 max_depth=3,
                 min_child_weight=1.5,
                 n_estimators=100000,                                                                    
                 reg_alpha=0.75,
                 reg_lambda=0.45,
                 subsample=0.6,
                 seed=42) 
```

```python
xgb1.fit(trainX,trainY)
```

```python
xgb1.base_score
```

```python
output = xgb1.predict(testX)
```

```python
sub['revenue'] = pd.DataFrame(output)
```

```python
sub.to_csv('submission.csv',index=False)
```
